```python
from functools import total_ordering
print("Welcome to the Simple Number Evaluation Program!")
print("How many numbers would you like to enter?")
num_entries_str = input()
try:
    num_entries = int(num_entries_str)
    entered_numbers = []
    for i in range(num_entries):
        number_input_str = input(f"Enter number {i+1} (between 1 and 676767676767): ")
        entered_numbers.append(int(number_input_str))
    print("You entered the following numbers:", entered_numbers)

    print("\n--- Evaluating Numbers ---")
    for number in entered_numbers:
        if number % 2 == 0:
            print(f"Number {number} is even.")
        else:
            print(f"Number {number} is odd.")

    if entered_numbers:
        total_sum = sum(entered_numbers)
        average = total_sum / len(entered_numbers)
        print(f"\n--- Summary ---")
        print(f"Total sum of entered numbers: {total_sum}")
        print(f"Average of entered numbers: {average:.2f}")
    else:
        print("\nNo numbers were entered to calculate total and average.")

except ValueError:
    print("Invalid input. Please enter a whole number for how many entries you want.")
    if 'number_input_str' in locals():
        try:
            num_to_check = int(number_input_str)
            if num_to_check % 2 == 0:
                print("The last entered number was even.")
            else:
                print("The last entered number was odd.")
        except ValueError:
            print("Could not determine parity for the last entered input.")
    else:
        print("No number was entered to determine parity.")
print("chicken butter chicken")
```
